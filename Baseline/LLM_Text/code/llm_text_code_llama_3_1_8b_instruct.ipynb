## Extract Input Length

```python
def prompt_length(text):
    if pd.isnull(text):
        return 0
    temp_inputs = tokenizer(text, return_tensors="pt")
    input_length = temp_inputs.input_ids.shape[1]
    return input_length
```

```python
# Apply to both zero-shot and cot prompts -> Cloud Logs
cloud_processed['zero-shot-question-length'] = cloud_processed['zero-shot-question'].apply(prompt_length)
cloud_processed['cot-question-length'] = cloud_processed['cot-question'].apply(prompt_length)
```

```python
# Apply to device_processed dataset -> Device logs
device_processed['zero-shot-question-length'] = device_processed['zero-shot-question'].apply(prompt_length)
device_processed['cot-question-length'] = device_processed['cot-question'].apply(prompt_length)
```

# Generate Response

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)  # move model to device once
model.eval()
```

### Zero-Shot Inference

```python
# Zero-Shot generated response
def generate_zero_shot_answer(row):
    try:
        # Set a fixed seed for reproducibility
        torch.manual_seed(42)
        if torch.cuda.is_available():
            torch.cuda.manual_seed_all(42)

        #tokenize the text
        temp_inputs = tokenizer(row['zero-shot-question'], return_tensors="pt")
        input_length = temp_inputs.input_ids.shape[1] # input length -> optional
        # Apply truncation only if needed
        if input_length > 2048:
            inputs = tokenizer(
                row['zero-shot-question'],
                return_tensors="pt",
                truncation=True,
                max_length=2048,
                padding=False
            ).to(device)
        else:
            inputs = temp_inputs.to(device)

        input_ids = inputs.input_ids
        attention_mask = inputs.attention_mask
        prompt_length = input_ids.shape[1]

        max_new_tokens = 768

          # Generate response
        with torch.no_grad():
            output = model.generate(
                input_ids=input_ids,
                attention_mask=attention_mask,
                max_new_tokens=max_new_tokens,
                eos_token_id=tokenizer.eos_token_id,
                pad_token_id=tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.6,
                top_p=0.9,
                repetition_penalty= 1.2
            )

        generated_tokens = output[0][prompt_length:]  # Remove prompt part
        answer = tokenizer.decode(generated_tokens, skip_special_tokens=True)
        return answer

    except Exception as e:
        print(f"Error while generating for question: {row['zero-shot-question'][:45]}")
        print("Error details:", e)
        return "[Error generating response]"
```

### CoT Inference

```python
# Chain of Thought generated response
def generate_cot_answer(row):
    try:
        # Set a fixed seed for reproducibility
        torch.manual_seed(42)
        if torch.cuda.is_available():
            torch.cuda.manual_seed_all(42)

        #tokenize the text
        temp_inputs = tokenizer(row['cot-question'], return_tensors="pt")
        input_length = temp_inputs.input_ids.shape[1] # input length -> optional

        # Apply truncation only if needed
        if input_length > 2048:
                inputs = tokenizer(
                row['cot-question'],
                return_tensors="pt",
                truncation=True,
                max_length=2048,
                padding=False
            ).to(device)
        else:
            inputs = temp_inputs.to(device)

        input_ids = inputs.input_ids
        attention_mask = inputs.attention_mask
        prompt_length = input_ids.shape[1]

        max_new_tokens = 768

        # Generate response
        with torch.no_grad():
            output = model.generate(
                input_ids=input_ids,
                attention_mask=attention_mask,
                max_new_tokens=max_new_tokens,
                eos_token_id=tokenizer.eos_token_id,
                pad_token_id=tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.6,
                top_p=0.9,
                repetition_penalty= 1.2
            )

        generated_tokens = output[0][prompt_length:]  # Remove prompt part
        answer = tokenizer.decode(generated_tokens, skip_special_tokens=True)
        return answer

    except Exception as e:
        print(f"Error while generating for COT question: {row['cot-question'][:45]}")
        print("Error details:", e)
        return "[Error generating COT response]"
```
